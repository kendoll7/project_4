```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
movie_data_df = pd.read_csv(
    "clean_csvs/complete_movie_list.csv",
    index_col="id")

# Display  data
movie_data_df.head(5)
```

# Logistic Regression Model

```python
new_df = movie_data_df[['budget','revenue','all_genres']]
new_df.head(10)
```

```python
new_df.dtypes
```

```python
new_df['all_genres'].value_counts()
```

```python
#Encode genre data
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
new_df['encoded_genres'] = labelencoder.fit_transform(new_df['all_genres'])
new_df
```

```python
# budget, encoded_genres = x
# revenue = y
# run logistic regression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = new_df.drop(["revenue", "all_genres"], axis=1)
y = new_df["revenue"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=1)

#scale data
scaler = StandardScaler()

#Fit the training data to Standard Scaler
X_scaler = scaler.fit(X_train)

#Transform the training data with scaler
X_train_scaled = X_scaler.transform(X_train)

#Transform the testing data with scaler
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

#Create a Logistic Regression model
lr_model = LogisticRegression(solver='lbfgs', max_iter=200, random_state=1)

#Fit the model using the training data
lr_model.fit(X_train_scaled, y_train)
LogisticRegression(random_state=1)
```

```python
#Display training and testing data scores
print(f"Training data score: {lr_model.score(X_train_scaled, y_train)}")
print(f"Testing data score: {lr_model.score(X_test_scaled, y_test)}")
```

```python
#Make predictions. 
predictions = lr_model.predict(X_test_scaled)
results = pd.DataFrame({"Predictions": predictions, "Actual": y_test}).reset_index(drop=True)
results.head()
```

```python
from sklearn.metrics import accuracy_score

# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

# Linear Regression Model

```python
#Remove movies with budget = 0
filtered_df = movie_data_df[movie_data_df['budget'] > 0]
filtered_df.head(5)
```

```python
cleaned_movie_df = pd.DataFrame(filtered_df['budget'])
cleaned_movie_df['revenue'] = filtered_df['revenue']
cleaned_movie_df.head(5)
```

```python
revenue_budget_plot = cleaned_movie_df.hvplot.scatter(
    x = 'budget',
    y = 'revenue',
    width = 800,
    height = 400,
    rot = 90)

revenue_budget_plot
```

```python
# Separate the y variable, the labels
y = cleaned_movie_df["revenue"]

# Separate the X variable, the features
x = cleaned_movie_df["budget"]
```

```python
# Reformat data of the independent variable X as a single-column array
x = cleaned_movie_df["budget"].values.reshape(-1, 1)

# Display sample data
x[:5]
```

```python
x.shape
```

```python
# Create a model with scikit-learn
model = LinearRegression()
```

```python
# Fit the data into the model
model.fit(x, y)
```

```python
# Display the slope
print(f"Model's slope: {model.coef_}")
```

```python
# Display the y-intercept
print(f"Model's y-intercept: {model.intercept_}")
```

```python
# Display the model's best fit line formula
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]}X")
```

```python
# Display the formula to predict the revenue for a movie with a budget of 14000000
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]} * 14000000")

# Predict the salary for a person with 7 years of experience
y_prediction = model.intercept_ + model.coef_[0] * 14000000

# Display the prediction
print(f"Predicted revenue for a movie with a budget of $14000000: ${y_prediction:.2f}")
```

```python
# Make predictions using the X set
predicted_y_values = model.predict(x)

# Create a copy of the original data
df_movie_predicted = cleaned_movie_df.copy()

# Add a column with the predicted salary values
df_movie_predicted["revenue_predicted"] = predicted_y_values

# Display sample data
df_movie_predicted.head()
```

```python
#Create a line plot of the predicted salary values
best_fit_line = df_movie_predicted.hvplot.line(
   x = "budget",
   y = "revenue_predicted",
   color = "blue")

best_fit_line
```

```python
# Superpose the original data and the best fit line
revenue_budget_plot * best_fit_line
```

```python
r_squared = model.score(x, y)
print(r_squared)
print(f"53.6% of the variance in movie revenue can be explained by the input variable of budget")
```
