```python
def process_production_companies(df, column):
    production_companies = pd.json_normalize(df[column])
    production_companies = production_companies.rename(columns={'name': f'production_company_{column}'})
    production_companies = production_companies.drop(columns='id', errors='ignore')
    return production_companies

production_companies_0 = process_production_companies(concat_production_companies_df, 0)
production_companies_1 = process_production_companies(concat_production_companies_df, 1)
production_companies_2 = process_production_companies(concat_production_companies_df, 2)
production_companies_3 = process_production_companies(concat_production_companies_df, 3)
production_companies_4 = process_production_companies(concat_production_companies_df, 4)
production_companies_5 = process_production_companies(concat_production_companies_df, 5)
production_companies_6 = process_production_companies(concat_production_companies_df, 6)
production_companies_7 = process_production_companies(concat_production_companies_df, 7)
production_companies_8 = process_production_companies(concat_production_companies_df, 8)
production_companies_9 = process_production_companies(concat_production_companies_df, 9)
production_companies_10 = process_production_companies(concat_production_companies_df, 10)
production_companies_11 = process_production_companies(concat_production_companies_df, 11)
production_companies_12 = process_production_companies(concat_production_companies_df, 12)
production_companies_13 = process_production_companies(concat_production_companies_df, 13)
production_companies_14 = process_production_companies(concat_production_companies_df, 14)
production_companies_15 = process_production_companies(concat_production_companies_df, 15)
production_companies_16 = process_production_companies(concat_production_companies_df, 16)


production_companies_16
```

```python
production_companies_dfs_list = []

for i in range(17):
    production_companies_df = process_production_companies(concat_production_companies_df, i)
    production_companies_dfs_list.append(production_companies_df)
    
production_companies_dfs_list
```
