```python
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

```python
movies_df = pd.read_csv('csvs/train.csv')
movies_df.head()
```

```python
movies_df.columns
```

```python
movies_df_clean = movies_df.drop(['homepage', 'imdb_id', 'original_language', 
                                  'overview', 'poster_path', 'production_countries', 'status',
                                 'crew', 'tagline'], axis=1)
movies_df_clean.head()
```

```python
movies_df_2 = pd.read_csv('csvs/test.csv')
movies_df_2.head()
```

```python
movies_df_2.columns
```

```python
movies_df_2_clean = movies_df_2.drop(['homepage', 'imdb_id', 'original_language', 
                                  'overview', 'poster_path', 'production_countries', 'status',
                                 'crew', 'tagline'], axis=1)
movies_df_2_clean.head()
```

```python
complete_movies_df_clean = pd.concat([movies_df_clean, movies_df_2_clean])
complete_movies_df_clean.head(10)
```

```python
complete_movies_df_clean.info()
```

```python
complete_movies_df_clean.rename(columns={"cast": "actors"})
```

```python
nan_counts = complete_movies_df_clean.isna().sum()
nan_counts
```

```python
complete_movies_df_clean_all = complete_movies_df_clean.dropna()
complete_movies_df_clean_all
```

```python
complete_movies_df_clean_all.info()
```

```python
complete_movies_df_clean.to_csv('clean_csvs/complete_movie_list.csv', index=False)
```
